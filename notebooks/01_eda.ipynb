```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
## data loading
DATA_PATH = "../data/raw/creditcard.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.shape
df["Class"].value_counts()
df["Class"].value_counts(normalize=True) * 100
```

```python
##visualization data
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=df)
plt.title("Transaction Class Distribution (0 = Genuine, 1 = Fraud)")
plt.xlabel("Transaction Class")
plt.ylabel("Count")
plt.show()
```

```python
##transaction amount 
plt.figure(figsize=(8,5))
sns.histplot(df[df["Class"] == 0]["Amount"], bins=50, color="green", label="Genuine", stat="density", kde=True)
sns.histplot(df[df["Class"] == 1]["Amount"], bins=50, color="red", label="Fraud", stat="density", kde=True)
plt.legend()
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Time"], bins=50)
plt.title("Transaction Time Distribution")
plt.xlabel("Time (seconds since first transaction)")
plt.ylabel("Count")
plt.show()
```
